### Analysis

A total of 249 mice who were identified with SCC tumors received treatment from a range of drug regimens and had their tumor development observed and measured over the course of 45 days.  After completetion of the study, a master data set was created by reconciling the tumor measurements with the metadata for each individual mouse in the study.  Each unique observation within the data set would be identified by the Timepoint of that specific tumor measurement and the Mouse ID of the test subject being measured.  Based upon this criteria, it was determined that the data for test subject 'g989' had duplicate observations and therefore, had to be omitted so as not to distort the analysis.

Once the master data set was was cleaned and reviewed, a summary of statistical measures was created to compare the results of each individual drug regimen based upon the measured tumor volume:

Based on the calculated statistics, the drug regimens Capomulin, Ramicane, Infubinol, and Ceftamin were identified as having the most promise due to producing the lowest average tumor volumes with relatively low degrees of variation as per their respective standard deviation, variance, and standard error.

To gauge the reliability of these statistics, the total number of unique observations for each drug regimen was plotted on a bar chart to observe any influence based upon sample size.  Capamulin and Ramicane clearly had the highest number of observations (which may explain why these drug regimens had the lowest standard error as per central limit theorem) but overall each drug regimen had a large enough sample size (>30) for central limit theorem to apply and allow us to make inferences on the population.




An additional check was made by charting the gender composition of the mice in the study using a pie chart.  The results showed roughly a 50/50 split between male and female mice which allows us to exclude the possibility of gender bias in the results.



Since Capomulin, Ramicane, Infubinol, and Ceftamin were identified as the most promising treatments within the set, we isolated the data for these treatments and focused specifically on the final tumor volumes measured at the end of the 45-day period.  By deriving an IQR for the final tumor volumes for each drug regimen, we calculated an upper and lower bounds to determine if any outliers existed within the data.  This approach identified one observation for Ifubinol as a potential outlier.  By charting the data into boxplots, we were able to visually see the one Ifubinol observation as an outlier.



To further assess the results of Capomulin specifically, we isolated data for a randomly selected test subject (Mouse ID# l509) being treated with the Capomulin regiment.  By charting the tumor volume against against the timepoint the measurements were taken, we were able to get a visual sense of how well the drug potentially impacted the subject's tumor volume over time.



Our analysis then proceeded to see if any relationship existed between tumor volume and body weight for the mice treated with Capomulin.  After calculating the average tumor volume and average weight measurement of each mouse within the Capomulin treatment group, these variables were charted into a scatter plot.



Overall, the study provides the following insights:

#### Pymaceutical's Capomulin does show promise, but Ramicane is a competitive alternative
When looking at the line chart for the measured tumor volumes for 'Mouse 1509,' we do see some periods of tumor growth but the overall tumor volume across the entire 45-day time period has been a downward trend. These results suggest that Capomulin may indeed be an effective treatment in decreasing the the volume of SCC tumors.  However, when compared with the other promising treatments in the study, we see that Ramicane is a competitive alternative treatment.  We can see this within the boxplots of the tumor volumes observed at the end of the 45-day study period - the range and median of the final tumor volumes are lower in Ramicane than they are for Capomulin.

#### Weight may be a contributing factor to the study's results
By charting the average tumor volume and average weight measurement of each mouse treated with Capomulin into a scatter plot, a relationship between the two variables becomes visually apparent.  By calculating the Pearson Correlation Coefficient, we see that there is a strong positive correlation of 0.84.  This introduces the possibility that mouse weight could be distorting our ability to assess the results of the study and the influence of each drug regimen.  For example, if mice treated with Ramicane have lower tumor volumes at the end of the 45 day period than mice treated with Capomulin, is it because Ramicane was more effective?  Or is it because the mice treated with Ramicane weighed less?  Or a mix of both?  Given this possibility, it may be more effective to assess the change in tumor volume rather than the final measured tumor volume.  Alternatively, another study could be conducted in the future that utilizes an approach at selecting mice in a way that eliminates potential weight bias.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_data_complete = pd.merge(study_results, mouse_metadata, on="Mouse ID")

# Display the data table for preview
study_data_complete.head()
```

```python
# Checking the number of mice.
study_data_complete['Mouse ID'].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicates = study_data_complete[study_data_complete.duplicated(subset=['Mouse ID', 'Timepoint'])]
duplicates['Mouse ID'].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID.
study_data_complete.loc[study_data_complete["Mouse ID"] == "g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
study_data_cleaned = study_data_complete[study_data_complete["Mouse ID"] != "g989"]
study_data_cleaned.head()
```

```python
# Checking the number of mice in the clean DataFrame.
study_data_cleaned["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
mean = study_data_cleaned.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
median = study_data_cleaned.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
variance = study_data_cleaned.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
std_dev = study_data_cleaned.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
sem = study_data_cleaned.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary DataFrame.
stat_summary = pd.DataFrame({
    "Mean Tumor Volume":mean,
    "Median Tumor Volume":median,
    "Tumor Volume Variance":variance,
    "Tumor Volume Std. Dev.":std_dev,
    "Tumor Volume Std. Err.": sem
})
stat_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
study_data_cleaned.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(["mean","median","var","std","sem"])
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
greatest_timepoint = study_data_cleaned.groupby("Mouse ID")["Timepoint"].max()
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume = pd.merge(greatest_timepoint, study_data_cleaned, on=["Mouse ID","Timepoint"], how="inner")
final_tumor_volume
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.
for drug in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    drug_results = final_tumor_volume.loc[(final_tumor_volume["Drug Regimen"] == drug)]["Tumor Volume (mm3)"]
    
    # add subset
    tumor_vol_data.append(drug_results)
    
    # Determine outliers using upper and lower bounds
    drug_results_df = pd.DataFrame(drug_results)
    quartiles = drug_results_df.quantile([0.25, 0.5, 0.75])
    lowerq = quartiles["Tumor Volume (mm3)"][0.25]
    upperq = quartiles["Tumor Volume (mm3)"][0.75]
    median = quartiles["Tumor Volume (mm3)"][0.5]
    iqr = upperq-lowerq
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)

    outliers = drug_results_df.loc[(drug_results_df["Tumor Volume (mm3)"] > upper_bound) | (drug_results_df["Tumor Volume (mm3)"] < lower_bound)]
    outlier_measurements = outliers["Tumor Volume (mm3)"]
    print(f"{drug}'s potential outliers: {outlier_measurements}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
x_axis = np.arange(0, len(treatments), 1) + 1
tick_locations = [value for value in x_axis]
fig, ax = plt.subplots()

flierprops = dict(markerfacecolor='red', markersize=15)
ax.boxplot(tumor_vol_data,flierprops=flierprops)
plt.xticks(tick_locations, treatments)
ax.set_ylabel("Final Tumor Volume (mm3)")
plt.tight_layout()
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
all_capomulin_mice = study_data_cleaned.loc[study_data_cleaned["Drug Regimen"]=="Capomulin"]
random_mouse = all_capomulin_mice["Mouse ID"].sample(1)
random_mouse_ID = random_mouse.values[0]
Capomulin_mouse = study_data_cleaned[study_data_cleaned["Mouse ID"]== random_mouse_ID]
plt.plot(Capomulin_mouse["Timepoint"],Capomulin_mouse["Tumor Volume (mm3)"])
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title(f"Capomulin treatment of mouse {random_mouse_ID}")
plt.tight_layout()
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_group = study_data_cleaned[study_data_cleaned["Drug Regimen"]=="Capomulin"]
mouse_weight = capomulin_group.groupby("Mouse ID")["Weight (g)"].mean()
avg_tumor_volume = capomulin_group.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
plt.scatter(mouse_weight,avg_tumor_volume)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.tight_layout()
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
print(f"The correlation between mouse weight and average observed tumor volume is {round(st.pearsonr(mouse_weight,avg_tumor_volume)[0],2)}")


plt.scatter(mouse_weight,avg_tumor_volume)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

from scipy.stats import linregress
(slope, intercept, rvalue, pvalue, stderr) = linregress(mouse_weight, avg_tumor_volume)
regress_values = mouse_weight * slope + intercept
plt.plot(mouse_weight, regress_values, color="red")

plt.tight_layout()
plt.show()
```
